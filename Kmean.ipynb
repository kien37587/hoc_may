```python
import matplotlib.pyplot as plt
x=[1,3,4,3,2]
y=[2,1,2,3,2]
plt.plot(x,y,'go')
plt.show()

def distance(x,y):
  sq1=(x[0]-y[0])*(x[0]-y[0])
  sq2=(x[1]-y[1])*(x[1]-y[1])
  return sq1+sq2
```

```python
#K=3,chon batky 3 gtri (x,y) vao list
C1=[int(input("Nhap x1: ")),int(input("Nhap Y1: "))]
C2=[int(input("Nhap x2: ")),int(input("Nhap Y2: "))]
C3=[int(input("Nhap x3: ")),int(input("Nhap Y3: "))]
#Ds cac cum
list1=[]
list2=[]
list3=[]

for j in range(2): #Vi gtri dau vao la (x,y)-> mang 2 chieu,nene xet 2 vong for
  for i in range(5):
    XX=[x[i],y[i]]
    d1=distance(XX,C1)
    d11=distance(XX,C2)
    d111=distance(XX,C3)
    if d1<d11 and d1<d111:
      list1.append(i)
    elif d11<d1 and d11<d111:
      list2.append(i)
    else:
      list3.append(i)
  tx1=tx2=tx3=ty1=ty2=ty3=0
  if len(list1)>0:
    for h in list1:
      tx1=tx1+x[h]
      ty1=ty1+y[h]
    C1=[tx1/len(list1),ty1/len(list1)]
    print("Diem cum c1: ",C1)
    if len(list2)>0:
      for h in list2:
        tx2=tx2+x[h]
        ty2=ty2+y[h]
      C2=[tx2/len(list2),ty2/len(list2)]
    print("Diem cum c2: ",C2)
    if len(list3)>0:
      for h in list3:
        tx3=tx3+x[h]
        ty3=ty3+y[h]
      C3=[tx3/len(list3),ty3/len(list3)]
    print("Diem cum c3: ",C3)
    list11=list1
    list22=list2
    list33=list3
    list1=[]
    list2=[]
    list3=[]
point=['A','B','C','D','E']
print("vang",list11)
print("bac",list22)
print("dong",list33)
```
