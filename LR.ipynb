```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([41,54,63,54,48,46,62,61,64,71])
y=np.array([1250, 1380, 1425, 1425, 1450, 1300, 1400, 1510, 1575, 1650])

x_mean=np.mean(x)
y_mean=np.mean(y)
print('gia tri x trung binh: ',x_mean)
print('gia tri y trung binh: ',y_mean)

tu=np.sum((x-x_mean)*(y-y_mean))
mau=np.sum((x-x_mean)**2)
b1=tu/mau
b0=y_mean - b1*x_mean
print('gia tri b1: ',b1)
print('gia tri b0: ',b0)

plt.scatter(x,y,color='blue',label='Du lieu')
x_line=np.linspace(min(x),max(x),100)
y_line=b0+b1*x_line
plt.plot(x_line,y_line,color='gray',label=f"Duong hoi quy: y={b0:.2f}+{b1:.2f}x")
try:
  testx=float(input("Nhap gtri x: "))
  predicted_y=b0+b1*testx
  print(f"Gia tri Y du doan la: {predicted_y:.2f}")
  plt.scatter(testx,predicted_y,color='cyan',label="Diem test")
except:
  print("Nhap lai")
plt.title("Bieu dien duong hoi quy")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
