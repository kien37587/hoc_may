```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'HS': [1, 2, 3, 4, 5],
    'Toan': [85, 60, 70, 70, 75],
    'TA': [80, 70, 50, 90, 85],
    'Tinh trang': ['Pass', 'Fail', 'Fail', 'Pass', 'Pass']
}
df = pd.DataFrame(data)

X=df[['Toan','TA']].values
group1=[]
group2=[]
for i in range (len(df)):
  for j in range (1):
    if (df.iloc[i,3]=='Pass'):
      h=(df.iloc[i,1]+df.iloc[i,2])
      group1.append(h)
    elif df.iloc[i,3]=='Fail':
      h=(df.iloc[i,1]+df.iloc[i,2])
      group2.append(h)
group1=[e for e in group1 if e]
group2=[e for e in group2 if e]

if group1:
  min_g1=group1[0]
  for i in range(len(group1)):
   if group1[i]<min_g1:
    min_g1=group1[i]
else:
   min_g1=0
if group2:
  max_g2=group2[0]
  for i in range (len(group2)):
    if group2[i]>max_g2:
      max_g2=group2[i]
else:
     max_g2=0
s=(min_g1+max_g2)/2

x1=int(input("Nhap diem toan: "))
x2=int(input("Nhap diem TA: "))

x0=np.array([0,s*2])
y0=np.array([s*2,0])
y=x1+x2-s
if y>0:
  print('Pass')
else:
  print('Fail')
```

```python
a1=[]
a2=[]
b1=[]
b2=[]
for i in range (len(df)):
  for j in range (1):
    if df.iloc[i,1]+df.iloc[i,2]-s>0:
      a1.append(df.iloc[i,1])
      a2.append(df.iloc[i,2])
    elif df.iloc[i,1]+df.iloc[i,2]-s<0:
      b1.append(df.iloc[i,1])
      b2.append(df.iloc[i,2])
plt.figure(figsize=(8,6))
plt.plot(a1, a2, 'yo', label="Pass")
plt.plot(b1, b2, 'bo', label="Fail")
plt.plot(x0, y0, 'g-', linewidth=2, label="Đường phân cách")
plt.plot(x1, x2, 'ro', markersize=8, label="Test")
plt.plot([0, max_g2], [0, max_g2], 'y--', label=f"Biên Fail (tổng = {max_g2})")
plt.plot([min_g1, 0], [0, min_g1], 'm--', label=f"Biên Pass (tổng = {min_g1})")

plt.xlim(0, 150)
plt.ylim(0, 150)
plt.xlabel('Điểm Toán')
plt.ylabel('Điểm TA')
plt.title('Phân loại Pass/Fail')
plt.legend(loc="best")
plt.grid(True, alpha=0.3)
plt.show()
```
