```python
tong_so_qua = 1000
so_luong_moi_qua = {'chuoi': 500, 'cam': 300, 'khac': 200}
thuoc_tinh = {
    'chuoi': [400, 100, 350, 150, 450, 50],
    'cam': [0, 300, 150, 150, 300, 0],
    'khac': [100, 100, 150, 50, 50, 150]
}
#tính P(A),xs tien nghiem
prior={}
for loai,sl in so_luong_moi_qua.items():
  prior[loai]=sl/tong_so_qua
#tính P(A),xs tien nghiem
prior={}
for loai,sl in so_luong_moi_qua.items():
  prior[loai]=sl/tong_so_qua
def likelihood(loai,tt,tt_chon):
  likelihood=1
  for tt_id in tt_chon:
    count=tt[tt_id-1]
    total=so_luong_moi_qua[loai]
    if total==0:
      return 0
    likelihood*=count/total
  return likelihood
def mle(tt_chon):
  post={}
  for loai,tt in thuoc_tinh.items():
      post[loai]=likelihood(loai,tt,tt_chon)*prior[loai]
  return post
def mle_normalized(tt_chon):
  post=mle(tt_chon)
  total=sum(post.values())
  if total==0:
    return {loai: 0 for loai in post}
  return {loai: value/total for loai,value in post.items()}
```

```python
while True:
  print("Nhập các thuộc tính của quả cần dự đoán:")
  print("1: Dài, 2: Không dài, 3: Ngọt, 4: Không ngọt, 5: Vàng, 6: Không vàng")
  thuoc_tinh_chon = list(map(int, input("Nhập các số (phân cách bằng dấu phẩy): ").split(',')))

  print("\nPosterior MLE: ")
  post_unormal=mle(thuoc_tinh_chon)
  for loai,value in post_unormal.items():
    print(f"{loai}: {value:.6f}")
  print("Du doan cua MLE: ",max(post_unormal,key=post_unormal.get))
  print("\nPosterior MLE_normalized: ")
  post_normal=mle_normalized(thuoc_tinh_chon)
  for loai,value in post_normal.items():
    print(f"{loai}: {value:.2%}")
  print("Du doan cua MLE_normalized: ",max(post_normal,key=post_normal.get))

  key=input("\nCo muon tiep tuc ko:    y:yes  n:no ")
  if key =="n":
    break
```
