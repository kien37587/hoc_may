```python
import pandas as pd
import numpy as np
tong_so_qua=1000
so_luong_moi_qua={'chuoi':500,'cam':300,'khac':200}
tong_thuoc_tinh=[500,500,650,350,800,200]
tt = {
    'chuoi': [400, 100, 350, 150, 450, 50],
    'cam': [0, 300, 150, 150, 300, 0],
    'khac': [100, 100, 150, 50, 50, 150]
}
#P(A)
prior={}
for loai,soluong in so_luong_moi_qua.items():
  prior[loai]=soluong/tong_so_qua
#P(Bi/A)
def likelihood(loai,tt,tt_chon):
  likelihood=1
  for tt_id in tt_chon:
    count=tt[loai][tt_id-1]
    total=so_luong_moi_qua[loai]
    if total==0:
      return 0
    likelihood*=count/total
  return likelihood
#P(A|Bi)
def nbc(tt_chon):
  post={}
  for loai,tt_id in tt.items():
    post[loai]=likelihood(loai,tt,tt_chon)*prior[loai]
  mau=1
  for tt_id in tt_chon:
    mau*=tong_thuoc_tinh[tt_id-1]/tong_so_qua
  for loai in post:
    post[loai]=post[loai]/mau
  return post
#nhap
while True:
  print("Nhập các thuộc tính của quả cần dự đoán:")
  print("1: Dài, 2: Không dài, 3: Ngọt, 4: Không ngọt, 5: Vàng, 6: Không vàng")
  tt_chon=list(map(int,input("Nhập các số (phân cách bằng dấu phẩy): ").split(',')))
  post=nbc(tt_chon)
  print("XS hậu nghiệm: ")
  for loai,xacsuat in post.items():
    print(f"{loai}: {xacsuat:.2%}")
  print("Dự đoán: ",max(post,key=post.get))
  key = input("\nBạn có muốn tiếp tục?  y:có   n:không \n")
  if key != 'y':
    break
```
