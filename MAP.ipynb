```python
from math import comb
import numpy as np

print("Nhap: ")
flips=[]
for i in range(1,11):
  while True:
    data=input(f"Nhap lan tung thu {i:2d}: ").strip()
    if data in['N','S']:
      flips.append(data)
      break
    else:
      print("Go lai dung yc: ")

num_flips = len(flips)
num_heads = flips.count('N')
num_tails = num_flips - num_heads

# Giả thuyết xác suất ngửa
p_h = [1/3, 1/2, 2/3]
b = list(set(p_h))
print(b)

NI = []
for i in b:
    ni = p_h.count(i)
    NI.append(ni)
print(NI)
posterior_probs = []
p1 = []  # Lưu P(D|H) - Likelihood

# BƯỚC 1: Tính Likelihood P(D|H) cho từng p trong b
print("\nTính Likelihood P(D|H):")
for p in b:
    likelihood = comb(num_flips, num_heads) * (p ** num_heads) * ((1 - p) ** num_tails)
    p1.append(likelihood)
    print(f" Voi p = {p:.3f} ,likelihood = {likelihood:.3f}")
print("\nTính Posterior P(H|D):")
unnormalized = []  # Lưu P(D|H) * P(H)
for i in range(len(b)):
    p = b[i]
    prior = NI[i] / len(p_h)           # P(H) = số đồng xu có p này / tổng
    unnorm = p1[i] * prior             # P(D|H) * P(H)
    unnormalized.append(unnorm)
# Tính P(D) = tổng các unnormalized
p_D = sum(unnormalized)
# Tính posterior = P(D|H) * P(H) / P(D)
for idx in range(len(b)):
    posterior = unnormalized[idx] / p_D
    posterior_probs.append(posterior)
    p = b[idx]
    print(f"  p = {p:.3f} → Posterior = {posterior:.6f}")
# Tìm giả thuyết có xác suất hậu nghiệm cao nhất
max_prob_index = posterior_probs.index(max(posterior_probs))
most_likely_coin = max_prob_index + 1
print(f"\n→ Đồng xu khả dĩ nhất: Đồng xu {most_likely_coin} (p = {b[max_prob_index]:.3f})")
print(f"→ Posterior: {max(posterior_probs):.3f}")
```
