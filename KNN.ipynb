```python
Bangdiem=[[85,80],[60,70],[70,50],[70,90],[75,85]]
outnew=[1,2,3,4,5]
output=[1,0,0,1,1]
Diem_minh=[int(input("Nhap diem thu 1 cua Minh: ")),int(input("Nhap diem thu 2: "))]
n=len(Bangdiem)
```

```python
import numpy as np
def kc(x,y):
  return np.power((x[0]-y[0]),2)+np.power((x[1]-y[1]),2)

Distance=[]
for i in range (n):
  Distance.append(kc(Diem_minh,Bangdiem[i]))

for i in range (0,n-1):
  for j in range(i+1,n):
    if(Distance[i]>Distance[j]):

      tam0=Distance[i]
      Distance[i]=Distance[j]
      Distance[j]=tam0

      tam1=output[i]
      output[i]=output[j]
      output[j]=tam1

      tam2=outnew[i]
      outnew[i]=outnew[j]
      outnew[j]=tam2

      tam3=Bangdiem[i]
      Bangdiem[i]=Bangdiem[j]
      Bangdiem[j]=tam3
print("Khoang cach sau khi sap xep la: \n",Distance)
print("Thu tu sau khi sap xep la: \n",outnew)
print("ds dau ra sau khi sap xep la:\n ",output)

dem0=0
dem1=0
k=[(int(input("Nhap so luong sinh vien muon so sanh: ")))]
K_value=k[0]
for i in range (K_value):
  diemsv=Bangdiem[i]
  kc1=Distance[i]
  ketqua="Do" if output[i]==1 else "truot"
  print(f"Sv stt{i}  KC: {kc1}  Diem: {diemsv},Ket qua: {ketqua}")
for i in range (len(k)):
  for j in range (k[i]):
    if output[j]==1:
      dem1=dem1+1
    else:
      dem0=dem0+1
  if dem1>dem0:
      print("So nhan do: ",dem1)
      print("So nhan truot: ",dem0)
      print("Voi K= ",k[i],"Thi minh do")
      dem0=dem1=0
  else:
      print("So nhan do: ",dem1)
      print("So nhan truot: ",dem0)
      print("Voi k= ",k[i],"Thi minh truot")
      dem0=dem1=0
```
